```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as sts

sns.set_context("poster")
sns.set_style("white")

%matplotlib inline
```

```python
data_dir = "../../Data/0_hvh/Clean"
data = pd.read_csv(data_dir + '/all_data_with_initials0.csv').drop('Unnamed: 0', axis=1)
```

```python
data.head()
```

```python
f, axes = plt.subplots(1, 1, figsize=(10,10), squeeze=False)

x = np.array([np.median(data.loc[data.subject == i, 'rt']) 
              for i in data.loc[data.subject%2 == 1, 'subject'].unique()])
y = np.array([np.median(data.loc[data.subject == i, 'rt']) 
              for i in data.loc[data.subject%2 == 0, 'subject'].unique()])

ax = axes[0, 0]
sns.regplot(x, y, ax=ax, ci=None, color='black', scatter_kws={'s':40})
ax.set_xlim([0,30])
ax.set_xlabel("First Player Median Response Time (s)", size=24)
ax.set_ylim([0, 30])
ax.set_ylabel("Second Player Median Response Time (s)", size=24)
ax.set_title("Median Response Correlation In Opponent Pairs", size=28, y=1.05)
m, b, r, p, err = sts.linregress(x,y)
f.text(x=.7, y=.82, s="r = " + str(r)[:4], size=24, family="Monospace")
f.text(x=.7, y=.79, s="p = " + str(p)[:6], size=24, family="Monospace")

sns.despine()
f.savefig('../../Figures/RT_Corr.png')
```

```python
f, axes = plt.subplots(1, 1, figsize=(12,12), squeeze=False)

x = np.array([np.median(data.loc[data.subject == i, 'rt']) 
              for i in data.loc[data.subject%2 == 1, 'subject'].unique()])
y = np.array([np.median(data.loc[data.subject == i, 'rt']) 
              for i in data.loc[data.subject%2 == 0, 'subject'].unique()])

m, b, r, p, err = sts.linregress(x,y)
xl = np.linspace(0, 30, 11)
yl = m*xl + b

ax = axes[0, 0]
ax.scatter(x, y, c='black')
ax.set_xlim([0,30])
ax.set_xlabel("First Player Median Response")
ax.set_ylim([0, 30])
ax.set_ylabel("Second Player Median Response")
ax.set_title("Median Response Correlation In Opponent Pairs")
f.text(x=.8, y=.8, s="r = " + str(r)[:4], size=24)
f.text(x=.8, y=.77, s="p  = " + str(p)[:6], size=24)
ax.plot(xl, yl, color='black')
sns.despine()
# f.savefig(figure_out + "response_correlation.png", format="png")
```
